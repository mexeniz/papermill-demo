```python
dataset_name = 'iris'
```

```python
import papermill as pm
import seaborn as sns

data = sns.load_dataset(dataset_name)
```

## Visualise dataset

```python
data.head()
```

```python
if dataset_name == 'iris':
    sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=data)
elif dataset_name == 'tips':
    sns.scatterplot(x='total_bill', y='tip', hue='sex', data=data)
elif dataset_name == 'titanic':
    sns.scatterplot(x='age', y='fare', hue='class', data=data)   
```

## Record values

```python
count = len(data)
data_cols = list(data.columns)

pm.record('count', count)
pm.record('data_cols', data_cols)
```
